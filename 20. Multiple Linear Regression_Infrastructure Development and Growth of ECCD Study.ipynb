# INFRASTRUCTURE DEVELOPMENT AND GROWTH OF EARLY CHILDHOOD DEVELOPMENT CENTRES

```python
#Study topic
#INFLUENCE OF INFRASTRUCTURE DEVELOPMENT ON GROWTH OF EARLY CHILDHOOD DEVELOPMENT CENTRES IN BONDO SUB-COUNTY, KENYA.
```

```python
#Objectives of the study
#i) To examine the influence of stakeholder’s attitude towards infrastructure development on growth of Early Childhood Development centres  in Bondo sub-county
#ii) To assess the influence of policy on infrastructure development on growth of Early Childhood Development centres   in Bondo sub-county
#iii)To determine the influence of financial allocation towards infrastructure development on growth of Early Childhood Development centres  in Bondo sub-county
#iv)To establish the influence of management of infrastructure development on growth of Early Childhood Development centres  in Bondo sub-county
```

```python
#Study Hypotheses
#i) H0:There is no significant relationship between stakeholder’s attitude towards infrastructure development and growth of Early Childhood Development centres  in Bondo sub-county
#ii)H0: There is no significant relationship between policy on infrastructure development and growth of Early Childhood Development centres   in Bondo sub-county
#iii)H0:There is no significant relationship between financial allocation towards infrastructure development and growth of Early Childhood Development centres  in Bondo sub-county 
#iv)H0:There is no significant relationship between management of infrastructure development and growth of Early Childhood Development centres  in Bondo sub-county
```

```python
#Splitting the data into train and test sets
from sklearn.model_selection import train_test_split
```

```python
#Splitting the data into train and test sets
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.80,test_size=0.20)
```

```python
#importing statsmodels libraries
import statsmodels.api as sm
import statsmodels.formula as smf
from statsmodels.formula.api import ols
```

```python
# Describing the model
mod= sm.OLS(y_train, X_train)
# Fitting the model
res = mod.fit()   
# Summarizing the model
print(res.summary())   
```

```python
#Hypothesis Testing 1
#1.H0:There is no significant relationship between stakeholder’s attitude towards infrastructure development and growth of Early Childhood Development centres  in Bondo sub-county
mod=ols('GrowthOfECDCentres ~ StakeholderAttitude',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```

```python
#Hypothesis Testing 2
#2. H0:There is no significant relationship between policy on infrastructure development and growth of Early Childhood Development centres   in Bondo sub-county
mod=ols('GrowthOfECDCentres ~PolicyOnInfrustructure',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```

```python
#Hypothesis Testing 3
#3. H0:There is no significant relationship between financial allocation towards infrastructure development and growth of Early Childhood Development centres  in Bondo sub-county 
mod=ols('GrowthOfECDCentres ~ FinancialAllocation',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```

```python
#Hypothesis Testing 4
#4. H0:There is no significant relationship between management of infrastructure and growth of Early Childhood Development centres  in Bondo sub-county 
mod=ols('GrowthOfECDCentres ~ ManagementOfInfrustructure',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```
