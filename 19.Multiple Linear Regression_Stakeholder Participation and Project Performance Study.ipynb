```python
#Objectives of the study
#i.To examine the influence of stakeholder participation in project formulation on the performance of end child marriage project in Homa Bay County 
#ii.To assess the influence of stakeholder participation in project financing on the performance of end child marriage project in Homa Bay County 
#iii.To determine the influence of stakeholder participation in project implementation on the performance of end child marriage project in Homa Bay County 
#iv.To find out the influence of stakeholders participation in project monitoring and evaluation on the performance of end child marriage project in Homa Bay County
```

```python
#Study Hypotheses
#i.H0:There is no relationship between stakeholder participation in project formulation on the performance of end child marriage project in Homa Bay County 
#ii.H0:There is no relationship between stakeholder participation in project financing on the performance of end child marriage project in Homa Bay County 
#iii.H0:There is no relationship between stakeholder participation in project implementation on the performance of end child marriage project in Homa Bay County 
#iv.H0:There is no relationship between stakeholders participation in project monitoring and evaluation on the performance of end child marriage project in Homa Bay County
```

```python
#Splitting the data into train and test sets
from sklearn.model_selection import train_test_split
```

```python
#Splitting the data into train and test sets
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.80,test_size=0.20)
```

```python
#importing statsmodels libraries
import statsmodels.api as sm
import statsmodels.formula as smf
from statsmodels.formula.api import ols
```

```python
# Describing the model
mod= sm.OLS(y_train, X_train)
# Fitting the model
res = mod.fit()   
# Summarizing the model
print(res.summary())   
```

```python
#Hypothesis Testing 1
#1.H0:There is no relationship between stakeholder participation in project formulation on the performance of end child marriage project in Homa Bay County
mod=ols('ProjPerformance ~ PartnInProjFormulation',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```

```python
#Hypothesis Testing 2
#2. H0:There is no relationship between stakeholder participation in project financing on the performance of end child marriage project in Homa Bay County 
mod=ols('ProjPerformance ~ PartnInProjFinancing',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```

```python
#Hypothesis Testing 3
#3. H0:There is no relationship between stakeholder participation in project implementation on the performance of end child marriage project in Homa Bay County
mod=ols('ProjPerformance ~ PartnInProjImplementation',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```

```python
#Hypothesis Testing 4
#4. H0:There is no relationship between stakeholders participation in project monitoring and evaluation on the performance of end child marriage project in Homa Bay County
mod=ols('ProjPerformance ~ PartnInProjMnE ',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```
