```python
#Study Hypotheses 
#i) H0:There is no significant relationship between promortional practices and product performance
#ii)H0:There is no significant relationship between pricing practices and product performance 
#iii)H0:There is no significant relationship between distribution practices and product performance
#iv)H0:There is no significant relationship between market research practices and product performance 
```

```python
#Importing the libraries
import pandas as pd
```

```python
#Importing the data
df=pd.read_csv('ProductPerformance.csv')
X=df.iloc[:,0:4]
y=df.iloc[:,4]
```

```python
#Getting the mean of each variable
df.mean()
```

```python
#Dealing with NaNs
df.fillna(df.mean())
```

```python
#Splitting the data into train and test sets
from sklearn.model_selection import train_test_split
```

```python
#Splitting the data into train and test sets
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.80,test_size=0.20)
```

```python
#importing statsmodels libraries
import statsmodels.api as sm
import statsmodels.formula as smf
from statsmodels.formula.api import ols
```

```python
# Describing the model
mod= sm.OLS(y_train, X_train)
# Fitting the model
res = mod.fit()   
# Summarizing the model
print(res.summary())   
```

```python
#Hypothesis Testing 1
#1.H0:There is no significant relationship between promortional practices and product performance
mod=ols('ProductPerformance ~ PromortionalPractices',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```

```python
#Hypothesis Testing 2
#2. H0:There is no significant relationship between pricing practices and product performance 
mod=ols('ProductPerformance ~ PricingPractices',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```

```python
#Hypothesis Testing 3
#3. H0:There is no significant relationship between distribution practices and product performance
mod=ols('ProductPerformance ~ DistributionPractices',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```

```python
#Hypothesis Testing 4
#4. H0:There is no significant relationship between market research practices and product performance
mod=ols('ProductPerformance ~ MarketResearchPractices',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```
