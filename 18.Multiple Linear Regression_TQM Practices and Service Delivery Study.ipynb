```python
#Objectives of the study
#i.To find out the influence of management commitment on service delivery in healthcare in public health facilities in Kisumu East Sub-county, Kenya.
#ii.To determine the influence of staff training on service delivery in healthcare in public health facilities in Kisumu East Sub-county, Kenya.
#iii.To examine the influence of employee involvement on service delivery in healthcare in public health facilities in Kisumu East Sub-county, Kenya.
#iv.To investigate the influence of continuous improvement on service delivery in healthcare in public health facilities in Kisumu East Sub-county, Kenya.
```

```python
#Study Hypotheses
#i.H0:There is no significant relationship between management commitment and service delivery in healthcare in public health facilities in Kisumu East Sub-county, Kenya.
#ii.H0:There is no significant relationship between staff training and service delivery in healthcare in public health facilities in Kisumu East Sub-county, Kenya.
#iii.H0:There is no significant relationship between employee involvement and service delivery in healthcare in public health facilities in Kisumu East Sub-county, Kenya.
#iv.H0:There is no significant relationship between continuous improvement and service delivery in healthcare in public health facilities in Kisumu East Sub-county, Kenya.
```

```python
#Splitting the data into train and test sets
from sklearn.model_selection import train_test_split
```

```python
#Splitting the data into train and test sets
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.80,test_size=0.20)
```

```python
#importing statsmodels libraries
import statsmodels.api as sm
import statsmodels.formula as smf
from statsmodels.formula.api import ols
```

```python
# Describing the model
mod= sm.OLS(y_train, X_train)
# Fitting the model
res = mod.fit()   
# Summarizing the model
print(res.summary())   
```

```python
#Hypothesis Testing 1
#1.H0:There is no significant relationship between management commitment and service delivery in healthcare in public health facilities in Kisumu East Sub-county, Kenya.
mod=ols('ServiceDelivery ~ ManagemntCommitment',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```

```python
#Hypothesis Testing 2
#2. H0:There is no significant relationship between staff training and service delivery in healthcare in public health facilities in Kisumu East Sub-county, Kenya. 
mod=ols('ServiceDelivery ~ StaffTraining',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```

```python
#Hypothesis Testing 3
#3. H0:There is no significant relationship between employee involvement  and service delivery in healthcare in public health facilities in Kisumu East Sub-county, Kenya.
mod=ols('ServiceDelivery ~ EmployeeInvolvement',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```

```python
#Hypothesis Testing 4
#4. H0:There is no significant relationship between continuous improvement and service delivery in healthcare in public health facilities in Kisumu East Sub-county, Kenya.
mod=ols('ServiceDelivery ~ ContinuousImprovement',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```
