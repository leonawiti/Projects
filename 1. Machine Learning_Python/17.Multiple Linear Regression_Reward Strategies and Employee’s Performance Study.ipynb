```python
#Objectives of the study
#1.To establish the extent to which pay as a component of reward strategies influences employee’s work performance within hospitals in Kisumu East Sub-county.
#2.To establish the extent to which job security as a component of reward strategies influences employee’s performance within hospitals in Kisumu East Sub-county.
#3.To establish the extent to which recognition as a component of reward strategy influences employee’s work performance within hospitals in Kisumu East Sub-county.
#4.To establish the extent to which promotion influences employee’s work performance within hospitals in Kisumu East Sub-county.
```

```python
#Study Hypothesis 
#1.H0:There is no relationship between pay as a component of reward strategies and employee’s work performance within private and public hospitals in Kisumu East Sub-county.
#2.H0:There is no relationship between job security as a component of reward strategies and employee’s performance within private and public hospitals in Kisumu East Sub-county.
#3.H0:There is no relationship between recognition as a component of reward strategy and employee’s work performance within private and public hospitals in Kisumu East Sub-county.
#4.H0:There is no relationship between promotion and employee’s work performance within private and public hospitals in Kisumu East Sub-county.
```

```python
#Splitting the data into train and test sets
from sklearn.model_selection import train_test_split
```

```python
#Splitting the data into train and test sets
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.80,test_size=0.20)
```

```python
#importing statsmodels libraries
import statsmodels.api as sm
import statsmodels.formula as smf
from statsmodels.formula.api import ols
```

```python
# Describing the model
mod= sm.OLS(y_train, X_train)
# Fitting the model
res = mod.fit()   
# Summarizing the model
print(res.summary())   
```

```python
#Hypothesis Testing 1
#1.H0:There is no relationship between pay as a component of reward strategies and employee’s work performance within private and public hospitals in Kisumu East Sub-county.
mod=ols('WorkPerformance ~ Pay',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```

```python
#Hypothesis Testing 2
#2. H0:There is no relationship between job security as a component of reward strategies and employee’s work performance within private and public hospitals in Kisumu East Sub-county.
mod=ols('WorkPerformance ~ JobSecurity',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```

```python
#Hypothesis Testing 3
#3. H0:There is no relationship between recognition as a component of reward strategies and employee’s work performance within private and public hospitals in Kisumu East Sub-county.
mod=ols('WorkPerformance ~ Recognition',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```

```python
#Hypothesis Testing 4
#4. H0:There is no relationship between promotion as a component of reward strategies and employee’s work performance within private and public hospitals in Kisumu East Sub-county.
mod=ols('WorkPerformance ~ Promotion',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```
