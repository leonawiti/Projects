```python
from sklearn.model_selection import train_test_split
#Spliting the data to train and test
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,test_size=0.2)
```

```python
#Fitting the regression model 
import statsmodels.api as sm
import statsmodels.formula as smf
from statsmodels.formula.api import ols
```

```python
# Describing the model
mod= sm.OLS(y_train, X_train)
# Fitting the model
res = mod.fit()   
# Summarizing the model
print(res.summary())   
```

```python
#Importing library
#Hypothesis Testing 1
#H0:There is no significant relationship between passive participation among stakeholders and sustainability of community development projects implemented by Plan International in Homa Bay
mod=ols('ProjectSustainability ~ PassiveParticipation',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```

```python
#Hypothesis Testing 2
#H0:There is no significant relationship between interactive participation among stakeholders and sustainability of community development projects implemented by Plan International in Homa Bay Town Sub-County
mod=ols('ProjectSustainability ~ InteractiveParticipation',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```

```python
#Hypothesis Testing 3
#H0:There is no significant relationship between functional participation among stakeholders and sustainability of community development projects implemented by Plan International in Homa Bay Town Sub-County
mod=ols('ProjectSustainability ~ FunctionalParticipation',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```

```python
#Hypothesis Testing 3
#H0.There is no significant relationship between gender diversity and employee performance in public universities in Kenya
mod=ols('ProjectSustainability ~ PassiveParticipation',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```

```python
#Hypothesis Testing 4
#H0:There is no significant relationship between optimum participation among stakeholders on sustainability of community development projects implemented by Plan International in Homa Bay Town Sub-County
mod=ols('ProjectSustainability ~ OptimalParticipation',data=df).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)
print(aov_table)
```
